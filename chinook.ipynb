List the percentage of all the tracks scold in USA and Arrange them in the  descending order of percentage

```python
%%sql
with 
Most_track as
(
select  taa.track_id,
        taa.genre,
        taa.name,
        COUNT(*)  as number_of_tracks_scold,
        (CAST(COUNT(*) as float) /1051)*100
         as percentage
from
    (
        select tg.track_id,
                   tg.genre,
                   ara.name 
        from 
            (
                select ar.name,
                       a.album_id 
                from 
                album a inner join artist ar on 
                ar.artist_id=a.artist_id
            )as ara 
        inner join 
            (
                select t.track_id,
                        g.name as genre,
                        t.album_id 
                from track t inner join genre g on
                g.genre_id=t.genre_id
            ) as tg on tg.album_id=ara.album_id 
    ) as taa 
inner join 
(
    select il.track_id 
    from invoice i 
        inner join 
        invoice_line il on 
        il.invoice_id=i.invoice_id 
    where billing_country="USA"
) as iil on 
iil.track_id=taa.track_id
group by taa.genre 
order by 4 DESC )
select * from MoSt_track as m
```

 List out which genres sell the most tracks in the USA, out of the four given genres "Hip-Hop","Punk","Pop","Blues" and order according to the percentage of number of tracks  scold in descending order

```python
%%sql
with 
Most_track as
(
select  taa.track_id,
        taa.genre,
        taa.name,
        COUNT(*)  as number_of_tracks_scold,
        (CAST(COUNT(*) as float) /1051)*100
         as percentage
from
    (
        select tg.track_id,
                   tg.genre,
                   ara.name 
        from 
            (
                select ar.name,
                       a.album_id 
                from 
                album a inner join artist ar on 
                ar.artist_id=a.artist_id
            )as ara 
        inner join 
            (
                select t.track_id,
                        g.name as genre,
                        t.album_id 
                from track t inner join genre g on
                g.genre_id=t.genre_id where genre in("Alternative & Punk","Blues","Pop","Hip Hop/Rap")
            ) as tg on tg.album_id=ara.album_id 
    ) as taa 
inner join 
(
    select il.track_id 
    from invoice i 
        inner join 
        invoice_line il on 
        il.invoice_id=i.invoice_id 
    where billing_country="USA"
) as iil on 
iil.track_id=taa.track_id
group by taa.genre 
order by 4 DESC )
select * from MoSt_track as m
```

List total number of customers,total value of sales,average value of sales per customer on purchases from different countries.

```python
%%sql
WITH
join1 as 
(
    select c.customer_id,
    c.first_name||" "||c.last_name as name,
    c.country,
    c.support_rep_id,
    sum(i.total) as total 
    from 
    customer c inner join invoice i 
    on c.customer_id=i.customer_id 
    group by c.first_name||" "||c.last_name 
),
number_of_cutomers1 as
(
    select j.*,
    COUNT(*) number_of_customers 
    from join1 j group by j.country
),
more_customers as
(
    select * 
    from number_of_cutomers1 as n 
    where number_of_customers>1
UNION
    select 
    p.customer_id,
    p.name,
    p.country,
    p.support_rep_id,
    SUM(p.total) as total,
    SUM(p.number_of_customers) as number_of_customers
    from number_of_cutomers1 as p  
    group by p.number_of_customers having p.number_of_customers=1
)
select m.country,
       m.number_of_customers,
       m.total as total_value_of_sales,
       m.total/m.number_of_customers as AVG_of_sales 
       from  more_customers m order by number_of_customers DESC
```

List the  top 10  artist name  used in the most playlists with playlist_name and COUNT of number_of_playlist order with respect to number_of_playlist in descending order 

```python
%%sql
WITH
artist_album as
(
    select ar.name as artist_name,
    a.album_id 
    from artist ar inner join album a 
    on ar.artist_id=a.artist_id
),
track_album as
(
    select aa.artist_name,
    t.track_id 
    from artist_album aa inner join track t
    on aa.album_id=t.album_id
),
playlist_playlist_track as 
(
    select p.name as playlist_name,
    pt.track_id 
    from playlist p inner join playlist_track pt
    on p.playlist_id=pt.playlist_id
)
select  ta.artist_name,
        ppt.playlist_name,
        COUNT(*) as number_of_playlist  
        from track_album ta inner join playlist_playlist_track ppt 
        on ta.track_id= ppt.track_id 
        group by artist_name
        order by 3 
        DESC limit 10
```

List number of tracks  which has been purchased vs  number of tracks not purchased

```python
%%sql
with 
ilt as

(
    select t.track_id,
    COUNT(*) as number_of_tracks 
    from invoice_line il inner join track t 
    on t.track_id=il.track_id 
    group by 1
),
not_Selected as
 (
     select t.track_id 
     from track t
 ),
except1  as
(
    select ns.track_id 
    from not_Selected ns
EXCEPT
    select ilt.track_id 
    from ilt
) ,
not_purchased as
(
    select COUNT(*) as not_purchased
    from except1 e
),
purchased as
(
    select COUNT(*) as number_of_tracks_purchased
    from ilt
)
select * from purchased,not_purchased
```

List  email, first name, last name, and Genre of all Rock Music listeners. Return your list ordered alphabetically by email address starting with A.

```python
%%sql
select cii.first_name,
       cii.last_name,
       cii.email,
       tg.genre
        from 
           (
              select il.track_id,
                      ci.first_name,
                      ci.last_name,
                      ci.email 
                from
                    (
                      select c.first_name,
                             c.last_name,
                             c.email,
                             i.invoice_id
                             from customer c inner join invoice i 
                              on c.customer_id=i.customer_id
                    )
                      as ci 
                      inner join invoice_line il 
                      on il.invoice_id=ci.invoice_id
           )as cii inner join 
           (
               Select g.name as genre,
                      t.track_id 
                      from track t inner join genre g 
                       on g.genre_id=t.genre_id
                       where g.name="Rock"
           )as tg 
        on tg.track_id= cii.track_id 
        group by cii.first_name,cii.last_name 
        order by cii.email 
```

List the customer who is  best in each country and  Check that best customer listen to Rock gener,then select those best customers with the money_spend,email,genre,Country 

```python
%%sql
select cii.customer_name,cii.email,tg.genre,cii.Each_country,MAX(cii.Money_spend_customer)
from

   (
    select ci.customer_name,ci.email,il.track_id,ci.Each_country,ci.Money_spend_customer
       from
    (
    select  c.first_name||c.last_name as customer_name,
        i.billing_country as Each_country,i.invoice_id,
        c.email,
        MAX(i.total) as Money_spend_customer
        from customer c inner join invoice i 
        on c.customer_id=i.customer_id
        group by i.billing_country,c.first_name||c.last_name
        order by MAX(i.total) DESC
    )as ci 
inner join invoice_line il 
on il.invoice_id= ci.invoice_id

)as cii inner join 
(
    select g.name as genre,
    t.track_id 
    from genre g inner join track t
    on t.genre_id=g.genre_id
)tg
on tg.track_id=cii.track_id 
group by cii.Each_country
having genre="Rock"
order by cii.email
```

List top 10 artist writing the rock music and count number of tracks each artist has written and order according to number of tracks that artist have in descending order

```python
%%sql
select ara.name as artist_name ,Count(*) as total_count_tracks  from (
    select ar.name,
       a.album_id 
      from artist ar inner join album a 
      on a.artist_id=ar.artist_id
    ) as ara
    inner join 
    (
        select t.track_id,
        t.album_id,
        g.name as genre 
        from track t inner join genre g
        on g.genre_id=t.genre_id 
        where g.name="Rock"
    )as tg 
    on tg.album_id=ara.album_id
    group by ara.name
    order by 2 DESC
    limit 10
```

List Top 10 artist who earned more according to the InvoiceLines

```python
%%sql
select ara.artist_name,SUM(ilt.unit_price*ilt.quantity) as Earned from
    (
    select ar.name as artist_name,
    a.album_id 
    from artist ar inner join album a 
    on a.artist_id=ar.artist_id
    ) as ara
    inner join
    (select il.track_id,
     il.unit_price,
     il.quantity,
     t.album_id
     from invoice_line il inner join track t
     on t.track_id=il.track_id
    ) as ilt on 
    ara.album_id=ilt.album_id
    group by ara.artist_name
    order by 2 DESC
    limit 10
    
```

List the Customer who has the  highest total invoice_Amount 

```python
%%sql
select ci.customer_id,
        ci.customer_name,
        SUM(aratil.unit_price*aratil.quantity) as customer_spend 
from
   (
       select c.customer_id,
        c.first_name||c.last_name as customer_name,
        i.invoice_id 
       from customer c inner join invoice i 
    on c.customer_id=i.customer_id
   ) as ci
inner join  
   (
     select ara.name,
       til.invoice_id,
       til.unit_price,
        til.quantity,
        til.track_id
        from 
         (select ar.name,
                 a.album_id 
                from artist ar inner join album a 
            on a.artist_id=ar.artist_id
         ) as ara
        inner join
        (
            select il.invoice_id,
                    il.unit_price,
                    il.quantity,
                    t.album_id, 
                    t.track_id
            from invoice_line il inner join track t 
            on t.track_id=il.track_id
        ) as til 
      on til.album_id=ara.album_id
    ) as aratil 
on aratil.invoice_id=ci.invoice_id
group by customer_id
order by SUM(aratil.unit_price*aratil.quantity) DESC
limit 1
```

List the number of songs does the customer_id 12 bought  base on genre  and ordered with respect to number_of_songs in descending order

```python
%%sql
select iiltg.name as genre,
       COUNT(*) as number_of_genre
       from customer c  
inner join
    (
     select iil.customer_id,
            tg.name 
        from
          (
           select i.invoice_id,
                  il.track_id,
                  i.customer_id 
                  from invoice i 
                   inner join 
                   invoice_line il 
                   on i.invoice_id=il.invoice_id
          ) as iil 
     inner join
        (
            select g.name,
                   t.track_id 
                   from track t 
                   inner join 
                   genre g 
                   on g.genre_id=t.genre_id
        )as tg
        on tg.track_id=iil.track_id
    )as iiltg
    on iiltg.customer_id=c.customer_id 
    where c.customer_id=12
    group by iiltg.name
    order by COUNT(*) DESC
```

List the total amount spent across genres whose employee_id=13

```python
%%sql
select iiltg.name as genre,
       SUM(iiltg.total) as COST_spend_genre
       from customer c  
inner join
    (
     select iil.customer_id,
            tg.name,
            iil.total
        from
          (
           select i.invoice_id,
                  il.track_id,
                  i.customer_id,
                  i.total
                  from invoice i 
                   inner join 
                   invoice_line il 
                   on i.invoice_id=il.invoice_id
          ) as iil 
     inner join
        (
            select g.name,
                   t.track_id 
                   from track t 
                   inner join 
                   genre g 
                   on g.genre_id=t.genre_id
        )as tg
        on tg.track_id=iil.track_id
    )as iiltg
    on iiltg.customer_id=c.customer_id 
    where c.customer_id=13
    group by iiltg.name
    order by 2 DESC
```
